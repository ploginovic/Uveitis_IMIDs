```python
import pandas as pd
import numpy as np
```

```python
hla_data = pd.read_csv("HLA_alleles_RA.csv")
```

```python
hla_data.drop(labels=[i for i in hla_data.columns if "Unnamed" in i], axis=1, inplace=True)
```

```python
hla_data
```

```python
hla_data["HLA_Allele"].replace("*", "_").replace(":", "")
```

```python
hla_data['HLA_Allele'] = hla_data['HLA_Allele'].str.strip(' ')
```

```python
allele_scores = {}
for index, row in hla_data.iterrows():
    #print(row.four_digit)
    for four_digit in row.four_digit.split(","):
        key = row.HLA_Allele.strip(" ").replace("HLA-", '') + four_digit.strip(" ").replace("*0", "_").replace("*", "_").replace(":", "")
        
        allele_scores.update({key:round(np.log(row.OR),2)})
```
